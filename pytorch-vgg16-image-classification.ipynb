**Import libraries**

```python
import torchvision.models as models
from PIL import Image
from torchvision import transforms
import torch
```

**Import VGG16**

```python
model = models.vgg16(pretrained=True)
```

**Import VGG16 label list**

```python
Label_path = "/content/drive/MyDrive/Colab Notebooks/Image-classification-vgg16-pytorch/vgg16-label-list.txt"
imagenetLabels = dict(enumerate(open(Label_path)))
```

**Read image**

```python
img = Image.open("/content/drive/MyDrive/Colab Notebooks/Image-classification-vgg16-pytorch/dog.jpg")
```

**Transform image**

```python
transform =   transforms.Compose([
        transforms.Resize((224, 224)),
        transforms.RandomHorizontalFlip(),
        transforms.ColorJitter(brightness=0.2, contrast=0.2, saturation=0.1, hue=0.1),
        transforms.RandomAffine(degrees=40, translate=None, scale=(1, 2), shear=15, resample=False, fillcolor=0),
        transforms.ToTensor(),
        transforms.Normalize((0.485, 0.456, 0.406), (0.229, 0.224, 0.225))
])

img_t = transform(img)

batch_t = torch.unsqueeze(img_t, 0)
```

**Set evaluation mode for model**

```python
model.eval()
```

**Test image on model**

```python
out = model(batch_t)
```

**Extract probability**

```python
probabilities = torch.nn.Softmax(dim=-1)(out)
```

**Print predected class label along with probability**

```python
(label, prob) = (imagenetLabels[probabilities.argmax().item()],probabilities.max().item())
print("Label: {}\nPercentage: {:.2f}%".format(label, prob * 100))
```
